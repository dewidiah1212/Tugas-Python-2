3. Dari data0.csv, lakukan perhitungan setiap kolomnya dan simpan perhitungan setiap kolomnya di parameter dataset baru pada kolom yang sama:
*   Rata-rata
*   Standar deviasi
*   Skewness
*   Median
*   Energi
*   Min
*   Max

```python
rata_rata = data.mean()
standar_deviasi = data.std()
skewness = data.apply(skew)
median = data.median()
energi = (data ** 2).sum()
nilai_min = data.min()
nilai_max = data.max()

parameter_dataset = pd.DataFrame({
    'Rata-rata': rata_rata,
    'Standar Deviasi': standar_deviasi,
    'Skewness': skewness,
    'Median': median,
    'Energi': energi,
    'Min': nilai_min,
    'Max': nilai_max
})

print(parameter_dataset)

parameter_dataset.to_csv('parameter_dataset.csv', index=False)
```

4. Plot sinyal diatas dari no. 2 dengan multiplot 2x3, sinyal yang diplot dapat dipilih dari 6 kolom secara bebas

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

axes = axes.flatten()

for i in range(len(axes)):
    random_column = np.random.choice(data.columns)

    axes[i].plot(data.index, data[random_column])
    axes[i].set_title(random_column)
    axes[i].set_xlabel('Index')
    axes[i].set_ylabel('Value')

plt.tight_layout()

plt.show()
```
